```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
from matplotlib.transforms import Bbox

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (11,7)

# Read in the data

df = pd.read_excel(r'C:\Users\shgre\Downloads\LEGO Table 2.xlsx')
```

```python
# Scatter plot with sets released each year

plt.scatter(x=df['year'], y=df['SetsReleased'])

plt.title('Sets Released Each Year')

plt.xlabel('Year')

plt.ylabel('Sets Released')

plt.show
```

```python
# Let's use seaborn to get a better look

sns.scatterplot(x='year', y='SetsReleased', data=df, hue='ParentTheme')

plt.title('Sets Released Each Year')

plt.xlabel('Year')

plt.ylabel('Sets Released')

plt.legend(title='Parent Theme', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)

plt.show
```
